# Data Cleaning - Cricket Dataset(cricketinfo.com)

```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings(action= 'ignore')
```

```python
df = pd.read_csv('Cricket .csv')
df
```

```python
df.rename(columns={'Mat': 'Matches_played', 'NO': 'Not_outs', 'Runs': 'Runs_scored',
                   'HS': 'Highest_inns_score','Ave': 'Batting_average',
                   'BF': 'Balls_faced','SR': 'Batiing_strike_rate'}, inplace= True)
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df[df['Player'].duplicated()]
```

```python
df[df['Player'].isin(['H Sutcliffe (ENG)', 'KC Sangakkara (SL'])]
```

```python
df = df.drop_duplicates(subset=['Player'], keep='first')
df
```

```python
df.duplicated().sum()
```

### Separating Span into End date and Start date

```python
df['Span'].str.split(pat='-')
```

```python
df['Start'] = df["Span"].str.split(pat='-').str[0]
```

```python
df['End'] = df['Span'].str.split(pat= '-').str[1]
```

```python
df.head()
```

```python
df = df.drop(columns='Span').head()
df
```

```python
df['Start'] =df['Start'].astype('int')
```

```python
df['End'] = df['End'].astype('int')
```

```python
df.info()
```

### Creating a new Column for Career length

```python
df['Career_length'] = abs(df['Start']-df['End'])
df.head()
```

### Splitting up country from player

```python
df['Player']
```

```python
df['Player'].str.split(pat= '(').str[0]
```

```python
(df['Player'].str.split(pat= '(').str[1]).str.split(pat=')').str[0]
```

```python
df['Country'] = (df['Player'].str.split(pat='(').str[1]).str.split(pat=')').str[0]
df.head()
```

```python
df['Player'] = df['Player'].str.split(pat='(').str[0]
df
```

### Changing features into appropriate Datatypes 

```python
df.info()
```

```python
df['Balls_faced'] = df['Balls_faced'].str.split(pat='+').str[0]
df.head()
```

```python
df['Balls_faced'] = df['Balls_faced'].str.split(pat='*').str[0]
df.head()
```

```python
df['Highest_inns_score'] = df['Highest_inns_score'].str.split(pat='*').str[0]
df.head()
```

```python
df = df.astype({'Matches_played': 'int', 'Highest_inns_score': 'int',
          'Balls_faced': 'int', 'Batiing_strike_rate': 'float'})
```

```python
df.info()
```

```python
# Average career length 
df['Career_length'].mean()
```

```python
# What is the avereage Batiing Strike rate of Players who Played over 10 years 
df[df['Career_length'] > 10]['Batiing_strike_rate'].mean()
```

```python
# Find the number of players who played before 1960
df[df['Start'] < 1960]['Player'].count()
```

```python
# Find the maximum Highest_inns_score by country
df.groupby(by='Country')['Highest_inns_score'].max().to_frame('Highest_Inn_Country')
```
