## Trend Analysis

```python
#df['date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df['month'] = df['Date'].dt.month
```

```python
df['Date'].dt.year
```

```python
df.head()
```

```python
monthly_revenue = df.groupby('month')['Total'].sum().sort_values()
monthly_revenue
```

```python
monthly_revenue.plot(x='Month', y='Total_revenue', kind='line', grid= True)
plt.show()
```

- According to the table above, it can be seen that the highest Revenue recorded in the first month. 

```python
df.groupby('Time')['Total'].max().sort_values(ascending= False)
```

```python
df.groupby('Time')['Total'].max().sort_values(ascending= False).plot(figsize= (12,4), marker= '.',grid= True)
```

- Sales can in other words be termed as sales therefore the maximum 'Total'(revenue) was used and it can be seen above 
  that highest sales(revenue) was recorded at 14:53:00    

## Revenue analysis

```python
df.groupby('Branch')['Total'].sum().sort_values(ascending= False)
```

```python
ax = df.groupby('Branch')['Total'].sum().sort_values(ascending= False).plot(kind='bar',figsize=(7,4),title='Total Sales By Branch',grid= True)

ax.set_xlabel('Branch')
ax.set_ylabel('Total Sales')

plt.show()
```

- Branch 'C' recorded the highest revenue followed by Branch 'A' and the Branch 'B'

```python
df.groupby('Customer type')['Total'].sum().sort_values(ascending=False)
```

```python
df.groupby('Customer type')['Total'].sum().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')
plt.title('Total Sales By Customer Type')
plt.show()
```

- Customer type 'Member' recorded the highest revenue as compared to Customer type 'Normal'  

```python
df.groupby('Gender')['Total'].sum().sort_values(ascending=False)
```

```python
df.groupby('Gender')['Total'].sum().sort_values(ascending=False).plot(kind= 'bar')
```

- Females contributed the highest revenue as compared to Males

```python
df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product line')['Total'].sum().sort_values(ascending=False).plot(kind='bar', grid=True)
plt.title('Product line Total Sales')
plt.xlabel('Product line')
plt.ylabel('Total Sales')
plt.show()
```

- Food and beverages formed the highest revenue in terms of Product line 

```python
df.groupby('Payment')['Total'].sum().sort_values(ascending=False)
```

```python
df.groupby('Payment')['Total'].sum().sort_values(ascending=False).plot.bar(color='green', figsize=(8,5))
plt.title('Total Sales by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Sales')
plt.show()
```

- Most people prefered to pay in cash followed by Ewallet and then Credit card

## Relationship Analysis

```python
df['Unit price'].corr(df['Quantity']).round(3)
```

- From the above correllation values, it can be seen that there is zero correlation between Unit price and Quantity 
  In other words the there is no visible relationship between them.

```python
df['cogs'].corr(df['Total']).round(3)
```

- From the above correllation values, it can be seen that there is a perfect correlation between Cost of goods and Revenue 
  This implies as the revenue goes up, the cost also goes up.

```python
df['Rating'].corr(df['Unit price']).round(3)
```

- From the above correllation values, it can be seen that there is zero correlation between Unit Ratings and Unit price
  In other words the there is no visible relationship between them.
